```python
from datasets import load_dataset
import evaluate
from tqdm import tqdm
```

```python
import pandas as pd
# Load dataset from HuggingFace manually via pandas
df = pd.read_parquet("https://huggingface.co/datasets/virattt/financial-qa-10K/resolve/main/data/train-00000-of-00001.parquet")
```

```python
df
```

```python
!pip install rouge_score bert_score
```

```python
# Load metrics
rouge = evaluate.load("rouge")
bleu = evaluate.load("bleu")
bertscore = evaluate.load("bertscore")
```

```python
# Prepare prediction and reference lists
predictions = []
references = []

# Evaluate on first 50 rows (for quick test)
sample_df = df.head(50)

for _, row in tqdm(sample_df.iterrows(), total=len(sample_df)):
    prompt = row["question"]
    ground_truth = row["answer"]

    # Generate prediction
    output = pipe(prompt, max_new_tokens=100, do_sample=False)[0]["generated_text"]

    # Clean and store
    predictions.append(output.strip())
    references.append(ground_truth.strip())
```

```python
# Compute BLEU
bleu_result = bleu.compute(predictions=predictions,references=references)
```

```python
# Compute ROUGE
rouge_result = rouge.compute(predictions=predictions, references=references)
```

```python
# Compute BERTScore
bertscore_result = bertscore.compute(predictions=predictions, references=references, lang="en")
```

```python
# Print metrics
print("\n📊 Evaluation Results:")
print(f"BLEU Score: {bleu_result['bleu']:.4f}")
print(f"ROUGE-1: {rouge_result['rouge1']:.4f}")
print(f"ROUGE-2: {rouge_result['rouge2']:.4f}")
print(f"ROUGE-L: {rouge_result['rougeL']:.4f}")
print(f"BERTScore F1: {sum(bertscore_result['f1']) / len(bertscore_result['f1']):.4f}")
```
